### R^2 statistics

- ** To evaluate the model, how good it is. for that we have R^2 statistics**

"R squared is used for simple linear Regression only & it is check the confidense level of the model."

- R square is  type of matric {in technical term}.
- R^2 statistics:- The R-Square statistics provides a measure of fit. It takes the form of a proportion the  proportion of variance explained- and so it always takes on a value between 0 & 1. In simple words, it represents how much of our data is  being explained by our model.

```python
# R^2 = (TSS-RSS)/TSS= 1-{RSS/TSS}

# RSS= sumation (y-y cap)^2

# TSS= sumation (y-y mean)^2

# print the R-Squared value for the  model
# lm.rsquared
```

### Adjusted R^2 statistics

** "As we increase the number of independent variables in our equation the R^2 increases as well. But that doesn't mean that the new independent variables have any correlation with the  output variables for that we have adjusted R^2 statistics-

##### Adjusted R^2     =   { 1 -  ( 1 - R^2) (N-1)}   /  N-P-1
where,
- R^2 = sample R-squared
- P = Number of predictors
- N = Total sample size.
